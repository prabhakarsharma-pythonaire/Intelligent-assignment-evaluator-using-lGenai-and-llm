```python
from PIL import Image

def combine_images(images, images_per_combined=3):
    """
    Combine every `images_per_combined` images from the input list into a single image.
    
    :param images: List of PIL.Image objects.
    :param images_per_combined: Number of images to combine into one.
    :return: List of combined PIL.Image objects.
    """
    combined_images = []
    for i in range(0, len(images), images_per_combined):
        # Get the current batch of images
        batch = images[i:i + images_per_combined]
        
        # Determine the maximum width and total height for vertical combining
        widths, heights = zip(*(img.size for img in batch))
        max_width = max(widths)
        total_height = sum(heights)
        
        # Create a new blank image with the determined dimensions
        combined_image = Image.new('RGB', (max_width, total_height))
        
        # Paste the images one below the other
        y_offset = 0
        for img in batch:
            combined_image.paste(img, (0, y_offset))
            y_offset += img.height
        
        # Append the combined image to the result list
        combined_images.append(combined_image)
    
    return combined_images


images=convert_from_path('input/IMG20240220101927.pdf')
    
    # Combine the images
combined_images = combine_images(images, 3)
    
    # Save or display the combined images





import time
start_time=time.time()
from dotenv import load_dotenv
from pdf2image import convert_from_path
import os
import google.generativeai as genai
from IPython.display import display
from IPython.display import Markdown
import PIL.Image
import matplotlib.pyplot as plt
image=PIL.Image.open('new.jpg')

load_dotenv()
api=None

if api==None:
    api=os.getenv('GEMINI_API_KEY')

genai.configure(api_key=api)
# image=convert_from_path('D:\Major project 4th year\input\IMG20240220101927.pdf')
model=genai.GenerativeModel('gemini-1.5-pro-latest')
for i in range(len(combined_images)):
    response=model.generate_content([combined_images[i],'return the handwritten text without manupulation from the image'],stream=False)
    
    print(response.text)

print(time.time()-start_time,'This is the time')
```
